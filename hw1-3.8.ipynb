```python
import numpy as np
import matplotlib.pyplot as plt
import time
import os
os.getcwd()
```

```python
f = np.loadtxt('millikan.txt')
x = f[:,0]
y = f[:,1]
```

```python
#part a
plt.plot(x,y,'ko')
plt.title('Millikan Data')
plt.xlabel('frequency (Hz)')
plt.ylabel('voltage (V)')
plt.show()
```

```python
#part b
E_x = np.sum(x)/len(x)
E_y = np.sum(y)/len(y)
E_xx = np.sum(x*x)/len(x)
E_xy = np.sum(x*y)/len(x)
m = (E_xy-E_x*E_y)/(E_xx-E_x**2)
c = (E_xx*E_y-E_x*E_xy)/(E_xx-E_x**2)
print('m = ', m, '    (slope)')
print('c = ', c, '    (intercept)')
```

```python
#part c
plt.plot(x, y, 'ko')
plt.plot(x, m*x+c, 'k-')
plt.title('Millikan Data')
plt.xlabel('frequency (Hz)')
plt.ylabel('voltage (V)')
plt.show()
```
